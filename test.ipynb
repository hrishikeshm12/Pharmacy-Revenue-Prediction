```python
import pandas as pd

def fetch_period_data(dataframe, start_day, period=7):
    """
    Fetch data for a specified period from the start day.

    Parameters:
    - dataframe: pd.DataFrame - The dataset containing a 'day' column
    - start_day: int - The day to start the period from
    - period: int - Number of days to include (e.g., 7, 15, 30)

    Returns:
    - pd.DataFrame - Filtered data for the specified period
    """
    # Filter rows where 'day' is within the specified period
    filtered_data = dataframe[(dataframe['day'] >= start_day) & (dataframe['day'] < start_day + period)]
    return filtered_data

# Example usage:
# Load the dataset
df = pd.read_csv("test_data.csv",sep="|")

# Fetch data for a 15-day period starting from day 1
data_15_days = fetch_period_data(df, start_day=63, period=15)

# Fetch data for a 30-day (1 month) period starting from day 1
data_30_days = fetch_period_data(df, start_day=63, period=30)

# Display the fetched data
print("15-Day Data:\n", data_15_days)
print("30-Day Data:\n", data_30_days)
```

```python
# Import necessary libraries
import joblib
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import get_custom_objects
from tensorflow.keras.layers import LeakyReLU
import h2o
h2o.init()

# Register the LeakyReLU activation function
get_custom_objects().update({'LeakyReLU': LeakyReLU})

# Load the models from disk
loaded_catboost = joblib.load("./models/Catboost_Model.joblib")
loaded_lightgbm = joblib.load("./models/LGBM_Model.joblib")
loaded_randomforest = joblib.load("./models/Tuned_Randomforest.joblib")
loaded_dnn = load_model('./models/DNN_Model.h5')
loaded_xgboost = joblib.load("./models/Tuned_XGBoost.joblib")
loaded_h2o_gbm = h2o.import_mojo("./models/GBM_5_AutoML_1_20241031_73853.zip")

def average_prediction(X_data):
    """
    Function to calculate the average prediction across multiple models for an unknown dataset.
    
    Parameters:
    - X_data: Feature matrix for prediction
    
    Returns:
    - y_pred_average: Average predicted values across all models
    - avg_revenue: Average predicted revenue value
    """
    
    # Get predictions from each model
    y_pred_dnn = loaded_dnn.predict(X_data).flatten()
    y_pred_catboost = loaded_catboost.predict(X_data)
    y_pred_lightgbm = loaded_lightgbm.predict(X_data)
    y_pred_randomforest = loaded_randomforest.predict(X_data).flatten()
    y_pred_xgboost = loaded_xgboost.predict(X_data)
    
    # Convert H2O predictions to numpy array
    h2o_frame = h2o.H2OFrame(X_data) 
    y_pred_h2o_gbm = h2o.as_list(loaded_h2o_gbm.predict(h2o_frame), use_pandas=False)[1:]  # skip header row
    y_pred_h2o_gbm = np.array(y_pred_h2o_gbm, dtype=np.float32).flatten()
    
    # Average predictions across all models
    y_pred_average = (y_pred_dnn + y_pred_catboost + y_pred_lightgbm + y_pred_randomforest + y_pred_xgboost + y_pred_h2o_gbm) / 6
    
    # Calculate average predicted revenue
    avg_revenue = np.mean(y_pred_average)
    
    print("Average Predicted Revenue:", avg_revenue)
    return y_pred_average, avg_revenue

# Example usage:
# Replace X_data with your actual data for prediction
# y_pred_avg, avg_revenue = average_prediction(X_data)
```

```python
import pandas as pd
import json

def preprocess_input(user_input: dict, grouping_json_path: str, aggregate_json_path: str, pid: str) -> pd.DataFrame:
    """
    Transforms user input into the format required for model prediction using reference data from JSON files.

    Parameters:
    - user_input (dict): Dictionary of user inputs with keys matching original column names.
    - grouping_json_path (str): Path to the JSON file containing grouping data.
    - aggregate_json_path (str): Path to the JSON file containing aggregated data.
    - pid (str): Product ID for which the data needs to be processed.

    Returns:
    - pd.DataFrame: Preprocessed data ready for model input.
    """
    # Load JSON data
    with open(grouping_json_path, 'r') as json_file:
        groupings = json.load(json_file)
    
    with open(aggregate_json_path, 'r') as json_file:
        aggregate_data = json.load(json_file)

    # Convert user input dictionary to DataFrame
    input_df = pd.DataFrame([user_input])

    # Use aggregated data for the given pid
    if pid in aggregate_data:
        input_df['total_clicks'] = aggregate_data[pid]['total_clicks']
        input_df['total_baskets'] = aggregate_data[pid]['total_baskets']
        input_df['total_orders'] = aggregate_data[pid]['total_orders']
    else:
        input_df['total_clicks'], input_df['total_baskets'], input_df['total_orders'] = 0, 0, 0


    # Action Ratios
    input_df['click_to_basket_ratio'] = input_df['total_clicks'] / (input_df['total_baskets'] + 1e-5)
    input_df['basket_to_order_ratio'] = input_df['total_baskets'] / (input_df['total_orders'] + 1e-5)
    input_df['click_to_order_ratio'] = input_df['total_clicks'] / (input_df['total_orders'] + 1e-5)

    # Competitor Price Features
    input_df['price_competitiveness'] = (input_df['price'] - input_df['competitorPrice']) / input_df['competitorPrice']
    input_df['price_rrp_ratio'] = input_df['price'] / input_df['rrp']
    input_df['competitor_undercut_flag'] = (input_df['price'] < input_df['competitorPrice']).astype(int)

    # Revenue Per Action Features
    input_df['revenue_per_click'] = input_df['revenue'] / (input_df['total_clicks'] + 1e-5)
    input_df['revenue_per_basket'] = input_df['revenue'] / (input_df['total_baskets'] + 1e-5)
    input_df['revenue_per_order'] = input_df['revenue'] / (input_df['total_orders'] + 1e-5)

    # Interaction Features
    input_df['price_availability_interaction'] = input_df['price'] * input_df['availability']
    input_df['competitor_availability_interaction'] = input_df['competitorPrice'] * input_df['availability']

    return input_df

# Example usage
user_input = {
    'group': 'Group A',
    'pharmForm': 'Form X',
    'manufacturer': 'Manufacturer Y',
    'category': 'Category Z',
    'unit': 'ML',
    'salesIndex': 40,
    'competitorPrice': 45,
    'price': 50,
    'availability': 1,
    'click': 1,
    'basket': 0,
    'order': 1,
    'revenue': 500,
    'rrp': 60
}

grouping_json_path = 'groupings.json'  # Path to your JSON file for groupings
aggregate_json_path = 'aggregate_data.json'  # Path to your JSON file for aggregated data
pid = 'example_pid'
preprocessed_data = preprocess_input(user_input, grouping_json_path, aggregate_json_path, pid)
print(preprocessed_data)
```

```python
import pandas as pd

def one_hot_encode(input_data, reference_data, cat_cols):
    unique_values = {}

    # Custom order for 'unit' column
    unit_order = ['CM', 'G', 'KG', 'L', 'M', 'ML', 'P', 'ST']
    
    # Gather unique values for each categorical column from reference data
    for col in cat_cols:
        if col == 'unit':
            unique_values[col] = unit_order
        else:
            unique_values[col] = list(set(reference_data[col]))
    
    # Create one-hot encoding for each value in the categorical columns
    one_hot_dict = {f"{col}_{value}": [] for col in cat_cols for value in unique_values[col]}
    
    # Fill the one-hot encoded vectors
    for item in input_data:
        one_hot_vector = {f"{col}_{value}": 0 for col in cat_cols for value in unique_values[col]}
        for col in cat_cols:
            if item[col] in unique_values[col]:
                one_hot_vector[f"{col}_{item[col]}"] = 1
        for key in one_hot_vector:
            one_hot_dict[key].append(one_hot_vector[key])
    
    return one_hot_dict

# Example usage
input_data = [
    {'unit': 'ML'}
]

# Assuming `df` is your DataFrame
reference_data = df

cat_cols = ['unit']

one_hot_encoded = one_hot_encode(input_data, reference_data, cat_cols)
print(one_hot_encoded)
```

```python
import pandas as pd
import json

def calculate_features_for_input(input_data, aggregate_data, groupings):
    pid = input_data['pid']
    
    # Aggregate Data Features
    if pid in aggregate_data:
        total_clicks = aggregate_data[pid]['total_clicks']
        total_baskets = aggregate_data[pid]['total_baskets']
        total_orders = aggregate_data[pid]['total_orders']
    else:
        total_clicks, total_baskets, total_orders = 0, 0, 0

    input_data['total_clicks'] = total_clicks
    input_data['total_baskets'] = total_baskets
    input_data['total_orders'] = total_orders

    # Action Ratios
    input_data['click_to_basket_ratio'] = round(total_clicks / (total_baskets + 1e-5), 2)
    input_data['basket_to_order_ratio'] = round(total_baskets / (total_orders + 1e-5), 2)
    input_data['click_to_order_ratio'] = round(total_clicks / (total_orders + 1e-5), 2)

    # Competitor Price Features
    input_data['price_competitiveness'] = round((input_data['price'] - input_data['competitorPrice']) / input_data['competitorPrice'], 2)
    input_data['price_rrp_ratio'] = round(input_data['price'] / input_data['rrp'], 2)
    input_data['competitor_undercut_flag'] = int(input_data['price'] < input_data['competitorPrice'])

    # Revenue Per Action Features
    input_data['revenue_per_click'] = round(input_data['revenue'] / (total_clicks + 1e-5), 2)
    input_data['revenue_per_basket'] = round(input_data['revenue'] / (total_baskets + 1e-5), 2)
    input_data['revenue_per_order'] = round(input_data['revenue'] / (total_orders + 1e-5), 2)

    # Interaction Features
    input_data['price_availability_interaction'] = round(input_data['price'] * input_data['availability'], 2)
    input_data['competitor_availability_interaction'] = round(input_data['competitorPrice'] * input_data['availability'], 2)

    # Adding Grouping Features
    for key, mapping in groupings.items():
        if key in input_data and input_data[key] in mapping:
            input_data[f'{key}_group'] = mapping[input_data[key]]
        else:
            input_data[f'{key}_group'] = None  # Default if key not found

    return input_data

# Load required JSON files
with open('aggregate_data.json', 'r') as agg_file:
    aggregate_data = json.load(agg_file)

with open('groupings.json', 'r') as group_file:
    groupings = json.load(group_file)

# Example Usage
user_input = {
    'pid': '6750',
    'group': '2FOI',  
    'pharmForm': 'TRO',
    'manufacturer': '255',
    'category': '193',
    'click': 1,
    'basket': 0,
    'order': 1,
    'price': 50,
    'competitorPrice': 45,
    'rrp': 60,
    'revenue': 500,
    'availability': 1
}

features = calculate_features_for_input(user_input, aggregate_data, groupings)
print(json.dumps(features, indent=4))
```

```python
import h2o
import json
import numpy as np
from h2o.frame import H2OFrame
import pandas as pd

# Initialize the H2O cluster
h2o.init()

# Load H2O model
model_path = "./models/GBM_5_AutoML_1_20241031_73853"  # Replace with the actual model path
h2o_model = h2o.load_model(model_path)

# Load JSON files for aggregate data and groupings
with open('aggregate_data.json', 'r') as agg_file:
    aggregate_data = json.load(agg_file)

with open('groupings.json', 'r') as group_file:
    groupings = json.load(group_file)

# Function to transform features
def transform_features(input_data, aggregate_data, groupings):
    pid = input_data['pid']
    
    # Aggregate Data Features (retrieve once)
    aggregate = aggregate_data.get(pid, {})
    total_clicks = aggregate.get('total_clicks', 0)
    total_baskets = aggregate.get('total_baskets', 0)
    total_orders = aggregate.get('total_orders', 0)
    

    input_data['total_clicks'] = total_clicks
    input_data['total_baskets'] = total_baskets
    input_data['total_orders'] = total_orders

    # Action Ratios
    input_data['click_to_basket_ratio'] = round(total_clicks / (total_baskets + 1e-5), 2)
    input_data['basket_to_order_ratio'] = round(total_baskets / (total_orders + 1e-5), 2)
    input_data['click_to_order_ratio'] = round(total_clicks / (total_orders + 1e-5), 2)

    # Competitor Price Features
    input_data['price_competitiveness'] = round((input_data['price'] - input_data['competitorPrice']) / input_data['competitorPrice'], 2)
    input_data['price_rrp_ratio'] = round(input_data['price'] / input_data['rrp'], 2)
    input_data['competitor_undercut_flag'] = int(input_data['price'] < input_data['competitorPrice'])

    # Revenue Per Action Features
    input_data['revenue_per_click'] = round(0.014797, 2)
    input_data['revenue_per_basket'] = round(0.025585, 2)
    input_data['revenue_per_order'] = round(0.011233, 2)

    # Interaction Features
    input_data['price_availability_interaction'] = round(input_data['price'] * input_data['availability'], 2)
    input_data['competitor_availability_interaction'] = round(input_data['competitorPrice'] * input_data['availability'], 2)

    # Adding Grouping Features
    for key, mapping in groupings.items():
        if key in input_data and input_data[key] in mapping:
            input_data[f'{key}_grouped'] = mapping[input_data[key]]
        else:
            input_data[f'{key}_grouped'] = None  # Default if key not found

    # Rearrange the output in the specified order
    ordered_output = {
        'adFlag': int(input_data.get('adFlag')),
        'availability': int(input_data.get('availability')),
        'competitorPrice': float(input_data.get('competitorPrice')),
        'price': float(input_data.get('price')),
        'content': float(input_data.get('content')),
        'genericProduct': int(input_data.get('genericProduct')),
        'rrp': float(input_data.get('rrp')),
        'group_grouped': pd.NA if input_data.get('group_grouped') is None else int(input_data.get('group_grouped')),
        'pharmForm_grouped': float(input_data.get('pharmForm_grouped', np.nan)),
        'manufacturer_grouped': pd.NA if input_data.get('manufacturer_grouped') is None else int(input_data.get('manufacturer_grouped')),
        'category_grouped': float(input_data.get('category_grouped', np.nan)),
        'total_clicks': int(input_data.get('total_clicks')),
        'total_baskets': int(input_data.get('total_baskets')),
        'total_orders': int(input_data.get('total_orders')),
        'click_to_basket_ratio': float(input_data.get('click_to_basket_ratio')),
        'basket_to_order_ratio': float(input_data.get('basket_to_order_ratio')),
        'click_to_order_ratio': float(input_data.get('click_to_order_ratio')),
        'price_competitiveness': float(input_data.get('price_competitiveness')),
        'price_rrp_ratio': float(input_data.get('price_rrp_ratio')),
        'competitor_undercut_flag': int(input_data.get('competitor_undercut_flag')),
        'revenue_per_click': float(input_data.get('revenue_per_click')),
        'revenue_per_basket': float(input_data.get('revenue_per_basket')),
        'revenue_per_order': float(input_data.get('revenue_per_order')),
        'price_availability_interaction': float(input_data.get('price_availability_interaction')),
        'competitor_availability_interaction': float(input_data.get('competitor_availability_interaction'))
    }
    ordered_output = pd.DataFrame([ordered_output])

    return ordered_output

# Sample user input
user_input = {
    'day': 1,
    'pid': '6570',
    'adFlag': 0,
    'availability': 2,
    'competitorPrice': 14.60,
    'click': 1,
    'basket': 0,
    'order': 0,
    'price': 16.89,
    'manufacturer': '255',
    'group': '2FOI',
    'content':50,
    'unit': "ML",
    'pharmForm': 'TRO',
    'genericProduct': 0,
    'salesIndex': 40,
    'category': "193.0",
    'rrp': 18.25
}

# Transform features
features = transform_features(user_input, aggregate_data, groupings)
print("Transformed Features:", features)

# Convert features to H2OFrame
h2o_features = H2OFrame(features)


# Predict using H2O model
predictions = h2o_model.predict(h2o_features)
y_pred_h2o_gbm = h2o.as_list(predictions, use_pandas=False)[1:]  # Skip header row
y_pred_h2o_gbm = np.array(y_pred_h2o_gbm, dtype=np.float32).flatten()

# Calculate average predicted revenue
avg_revenue = round(np.mean(y_pred_h2o_gbm), 2)

# Display results
print("Predicted Revenue:", y_pred_h2o_gbm)
print("Average Predicted Revenue:", avg_revenue)
```

```python
from flask import Flask, render_template, request, jsonify
import json
import h2o
import numpy as np
from h2o.frame import H2OFrame
import pandas as pd
from flask_cors import CORS
import requests

# Load JSON files for aggregate data and groupings
with open('aggregate_data.json', 'r') as agg_file:
    aggregate_data = json.load(agg_file)

with open('groupings.json', 'r') as group_file:
    groupings = json.load(group_file)

app = Flask(__name__)

# Function to transform features
def transform_features(input_data, aggregate_data, groupings):
    pid = input_data['pid']

    # Convert numeric fields to the correct type
    input_data['price'] = float(input_data['price'])
    input_data['competitorPrice'] = float(input_data['competitorPrice'])
    input_data['content'] = float(input_data['content'])
    input_data['rrp'] = float(input_data['rrp'])
    input_data['availability'] = int(input_data['availability'])
    input_data['genericProduct'] = int(input_data['genericProduct'])

    # Aggregate Data Features (retrieve once)
    aggregate = aggregate_data.get(pid, {})
    total_clicks = aggregate.get('total_clicks', 0)
    total_baskets = aggregate.get('total_baskets', 0)
    total_orders = aggregate.get('total_orders', 0)
    
    input_data['total_clicks'] = total_clicks
    input_data['total_baskets'] = total_baskets
    input_data['total_orders'] = total_orders

    # Action Ratios
    input_data['click_to_basket_ratio'] = round(total_clicks / (total_baskets + 1e-5), 2)
    input_data['basket_to_order_ratio'] = round(total_baskets / (total_orders + 1e-5), 2)
    input_data['click_to_order_ratio'] = round(total_clicks / (total_orders + 1e-5), 2)

    # Competitor Price Features
    input_data['price_competitiveness'] = round((input_data['price'] - input_data['competitorPrice']) / input_data['competitorPrice'], 2)
    input_data['price_rrp_ratio'] = round(input_data['price'] / input_data['rrp'], 2)
    input_data['competitor_undercut_flag'] = int(input_data['price'] < input_data['competitorPrice'])

    # Revenue Per Action Features
    input_data['revenue_per_click'] = round(0.014797, 2)
    input_data['revenue_per_basket'] = round(0.025585, 2)
    input_data['revenue_per_order'] = round(0.011233, 2)

    # Interaction Features
    input_data['price_availability_interaction'] = round(input_data['price'] * input_data['availability'], 2)
    input_data['competitor_availability_interaction'] = round(input_data['competitorPrice'] * input_data['availability'], 2)

    # Adding Grouping Features
    for key, mapping in groupings.items():
        if key in input_data and input_data[key] in mapping:
            input_data[f'{key}_grouped'] = mapping[input_data[key]]
        else:
            input_data[f'{key}_grouped'] = None  # Default if key not found

    # Rearrange the output in the specified order
    ordered_output = {
        'adFlag': int(input_data.get('adFlag')),
        'availability': int(input_data.get('availability')),
        'competitorPrice': float(input_data.get('competitorPrice')),
        'price': float(input_data.get('price')),
        'content': float(input_data.get('content')),
        'genericProduct': int(input_data.get('genericProduct')),
        'rrp': float(input_data.get('rrp')),
        'group_grouped': pd.NA if input_data.get('group_grouped') is None else int(input_data.get('group_grouped')),
        'pharmForm_grouped': float(input_data.get('pharmForm_grouped', np.nan)),
        'manufacturer_grouped': pd.NA if input_data.get('manufacturer_grouped') is None else int(input_data.get('manufacturer_grouped')),
        'category_grouped': float(input_data.get('category_grouped', np.nan)),
        'total_clicks': int(input_data.get('total_clicks')),
        'total_baskets': int(input_data.get('total_baskets')),
        'total_orders': int(input_data.get('total_orders')),
        'click_to_basket_ratio': float(input_data.get('click_to_basket_ratio')),
        'basket_to_order_ratio': float(input_data.get('basket_to_order_ratio')),
        'click_to_order_ratio': float(input_data.get('click_to_order_ratio')),
        'price_competitiveness': float(input_data.get('price_competitiveness')),
        'price_rrp_ratio': float(input_data.get('price_rrp_ratio')),
        'competitor_undercut_flag': int(input_data.get('competitor_undercut_flag')),
        'revenue_per_click': float(input_data.get('revenue_per_click')),
        'revenue_per_basket': float(input_data.get('revenue_per_basket')),
        'revenue_per_order': float(input_data.get('revenue_per_order')),
        'price_availability_interaction': float(input_data.get('price_availability_interaction')),
        'competitor_availability_interaction': float(input_data.get('competitor_availability_interaction'))
    }
    ordered_output = pd.DataFrame([ordered_output])

    return ordered_output

data = {
    'pid': '7403',
    'adFlag': '1',
    'availability': '50',
    'competitorPrice': '25.9',
    'price': '40.0',
    'content': '50',
    'rrp': '22.3',
    'manufacturer': '255',
    'group': '2FOI',
    'unit': 'KG',
    'pharmForm': 'TAB',
    'genericProduct': '1',
    'salesIndex': '40',
    'category': '50.0'
}

url = 'http://localhost:8080/predict'

final_data=transform_features(data,aggregate_data,groupings)


# Convert the DataFrame to a dictionary without index
ordered_output_dict = final_data.iloc[0].to_dict()

# Now create the CSV-like string format
input_data = ','.join([f"{key}:{value}" for key, value in ordered_output_dict.items()])
print(input_data)

response = requests.post(url, data=input_data)
prediction = response.json()

print(f"Predicted Value: {prediction}")
```
